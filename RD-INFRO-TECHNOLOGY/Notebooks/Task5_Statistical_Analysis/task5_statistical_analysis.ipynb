```python
import pandas as pd

df = pd.read_csv("../Dataset/NIFTY50_TRANSFORMED.csv")
df.describe()
```

```python
from scipy.stats import ttest_ind

positive_returns = df[df['Daily_Return_%'] > 0]['Daily_Return_%']
negative_returns = df[df['Daily_Return_%'] < 0]['Daily_Return_%']

t_stat, p_value = ttest_ind(positive_returns, negative_returns)

t_stat, p_value
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

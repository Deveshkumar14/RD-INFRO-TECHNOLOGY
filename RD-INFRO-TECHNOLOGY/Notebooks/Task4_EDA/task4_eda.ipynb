```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../Dataset/NIFTY50_TRANSFORMED.csv")
df.head()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Close'])
plt.title("NIFTY 50 Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df['Daily_Return_%'].dropna(), bins=50)
plt.title("Distribution of Daily Returns")
plt.xlabel("Daily Return (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['MA_20'], label='20-Day Moving Average')
plt.title("NIFTY 50 Close Price vs 20-Day Moving Average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

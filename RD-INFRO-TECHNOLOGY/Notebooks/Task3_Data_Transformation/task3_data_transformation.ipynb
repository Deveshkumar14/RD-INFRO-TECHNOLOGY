```python
import pandas as pd

df = pd.read_csv("../Dataset/NIFTY50_CLEANED.csv")
df.head()
```

```python
df['Daily_Return_%'] = df['Close'].pct_change() * 100
df.head()
```

```python
df[['Date', 'Close', 'Daily_Return_%']].head()
```

```python
df['MA_20'] = df['Close'].rolling(window=20).mean()
df.head(25)
```

```python
df.to_csv("../Dataset/NIFTY50_TRANSFORMED.csv", index=False)
```
